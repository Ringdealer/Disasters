```python
import dateparser
import numpy as np
import pandas as pd
import re

def strip_string(mod_df, col, pattern):
    i = 0
    for word in mod_df[col]:
        if isinstance(word, str):
            word = word.rstrip()
            match = re.search(pattern, word)
            if match:
                word = word.replace(match.string, '')
                mod_df.at[i, col] = word
                i += 1
            else:
                i += 1
        else:
            i += 1            
    return mod_df

def convert_to_isodate(df, col):
    i = 0
    for word in df[col]:        
        if isinstance(word, str):
            word = word.rstrip()
            try:
                df.at[i, 'ISO Date'] = dateparser.parse(word).isoformat()
                i += 1
            except(AttributeError):
                pass
                i += 1
        else:
            i += 1 
    return df


def create_new_col(df, pattern, src_col, new_col):
    i = 0
    for word in df[src_col]:        
        if isinstance(word, str):
            if(re.search(pattern,word)):
                match = re.search(pattern, word)
                df.at[i,new_col] = match.group()
                i += 1
            else:
                df.at[i,new_col] = ''
                i += 1
        else:
                df.at[i,new_col] = ''
                i += 1               
    return df


def convert_to_string(mod_df, col):
    mod_df[col] = mod_df[col].fillna('-1')
    mod_df[col] = mod_df[col].astype(str)
    mod_df[col] = mod_df[col].replace('-1', np.nan)    
    return mod_df
    

def str_to_nan(mod_df, column):
    for word in mod_df[column]:
        if isinstance(word, str):
            mod_df[column].replace(word, np.nan, inplace=True)
    return mod_df


def convert_to_int(mod_df, col):
    i = 0
    for word in mod_df[col]:
        if isinstance(word, str):
            word = word.rstrip()
            word = word.replace(',', '')
            if word.isdigit():
                mod_df.at[i, col] = int(word)
                i += 1
            else:
                i += 1
        else:
            i += 1
    return mod_df


def strip_symbol(mod_df, col, pattern):
    i = 0
    for word in mod_df[col]:
        if isinstance(word, str):
            word = word.replace(pattern, '')
            mod_df.at[i, col] = word
            i += 1
        else:
            i += 1
    return mod_df

def emend_values(orig_df, mod_df, source_col, target_col, pattern):
    i = 0
    for word in orig_df[source_col]:
        if isinstance(word, str):
            word = word.rstrip() 
            match = re.search(pattern, word)
            if match:
                 mod_df.at[i, target_col] = match.string
                 i += 1
            else:
                i += 1               
        else:
            i += 1    

    return mod_df



def convert_to_int(mod_df, col):
    """ Method for converting to strings to integer """
    i = 0
    for word in mod_df[col]:
        if isinstance(word, str):
            word = word.replace(',', '')
            if word.isdigit():
                mod_df.at[i, col] = int(word)
                i += 1
            else:
                i += 1
        else:
            i += 1
    return mod_df


def convert_to_nan(df, column):
    """
	Method for converting values to NaN
	"""
    i = 0
    for word in df[column]:
        if isinstance(word, str):
            df.at[i, column] = None
        i += 1

    return df


def fill_na(df, column):
    """
	Method for filling empty cells with NaN
	"""
    i = 0
    for word in df[column]:
        if word is None:
            df.at[i, column] = 0
        i += 1

    return df




```
